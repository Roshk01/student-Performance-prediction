# Data Preprocessing
encode catagorical data

```python
df_rename['result'] = ((df_rename['math score'] + df_rename['reading score'] + df_rename['writing score']) /300) * 100
```

```python
df_rename.head()
```

```python
df_rename = pd.get_dummies(df_rename, columns=[
    'gender',
    'ethinic',
    'parent_education',
    'lunch',
    'preparation'
], drop_first=True)
```

```python
df_rename['result'] = df_rename['result'].apply(lambda x: 1 if x > 50 else 0)
df_rename.sample(6)
```

```python
final_data = df_rename.rename(columns={'gender_male':'gender', 'ethinic_group B':'group_B','ethinic_group C':'group_C','ethinic_group D':'group_D','ethinic_group E':'group_E','preparation_none':'preparation'})
```

```python
final_data['result'].value_counts()
```

```python
final_data.sample(10)
```
